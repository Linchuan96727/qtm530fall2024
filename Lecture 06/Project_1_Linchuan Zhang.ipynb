```python
# Initialize an empty list to store data
data = []
```

```python
# Loop to iterate through the pages
for page in range(1, 7):  # Assuming there are 6 pages
    # Locate the element containing the product information
    search_1 = driver.find_element('xpath', '//div[@class="variant-stack-layout"]')
    
    # Get the outer HTML of this element
    html_code = search_1.get_attribute("outerHTML")
    
    # Use BeautifulSoup to parse the HTML content
    soup = BeautifulSoup(html_code, "html.parser")
    
    # Find all product titles (using h3 tag and class ps-title)
    product_titles = soup.find_all('h3', class_='ps-title')
    
    # Loop through the titles and extract the text, then append to the data list
    for title in product_titles:
        product_name = title.get_text(strip=True)
        data.append({"product_name": product_name})

    # Check if "Next" button exists and is enabled
    try:
        next_button = driver.find_element('xpath', '//button[@class="dds__button dds__button--tertiary dds__button--sm dds__pagination__next-page"]')
        next_button.click()  # Click the "Next" button to go to the next page
        time.sleep(2)  # Add a short wait to ensure the page loads
    except Exception as e:
        print("No more pages to load or encountered an issue:", e)
        break

# Convert the list of dictionaries into a DataFrame
data_df = pd.DataFrame(data)
```

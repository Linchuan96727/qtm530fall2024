(c) Split a dataset into subsets

<font size = "3">

- Create ``` list_status = [0,1]```
- Create ``` list_prob = [0.5, 0.5]```
- Create a new variable $n = 1000$ with the number of rows of the dataset
- Use 
```python
    np.random.choice(list_status,
                     size = n,
                     p = list_prob)
```
$\qquad$ to create a new column called "status" <br>
$\qquad$ which randomly assigns each row to "0" or "1" <br>
$\qquad$ with equal probability
- Use .query() to subset all observations with status == 0.<br>
store it in a new dataset called "data0"
- Use .query() to subset all observations with status == 1.<br>
store it in a new dataset called "data1"



````python
import numpy as np

np.random.seed(888)

list_status = [0,1]
list_prob = [0.5, 0.5]
n = 1000

data = {"status": np.random.choice(list_status, 
                                   size = n, 
                                   p = list_prob)}
data = pd.DataFrame(data)

# Subset all observations with status == 0
data0 = data.query("status == 0")
# Subset all observations with status == 1
data1 = data.query("status == 1")
````

(j) Plot by category

<font size = "3">

- Read "data_raw/wdi_urban.csv"<br>
This contains data from 1980 and 2020

- Create a list with the unique values of "year"
- Create a plot with a ...
    - Histogram of "prop_urbanpopulation" <br>
    for each year
    - Set the histogram option ``` alpha = 0.5 ``` for transparency
    - All the histograms should be overlapping
    - Create a legend with the list of years
    - Label the axes and the title

What do we learn from the graph?

HINT: See the example at the end of Lecture 10


````python
# Read "data_raw/wdi_urban.csv"
wdi_urban = pd.read_csv("data_raw/wdi_urban.csv")

# Create a list with the unique values of "year"
unique_years = wdi_urban["year"].unique()

# Creat the Histogram of "prop_urbanpopulation" for each year
for year in unique_years:
    df_year = wdi_urban.query("year == @year")
    plt.hist(df_year["prop_urbanpopulation"], alpha=0.5)

plt.legend(labels = unique_years)
plt.xlabel("Proportion of Urban Population")
plt.ylabel("Frequency")
plt.title("Histogram of Urban Population Proportion between 1980 and 2020")
plt.show()
````

````python
# What do we learn from the graph?

# 1. The distribution of urban population proportions across countries in 1980 is different from it in 2020.

# 2. From 1980 to 2020, there is an increase in the number of countries with relatively higher urban population proportions (for example, over 50%),
#  and a decrease in the number of countries with relatively lower urban population proportions.

# 3. Substantively, in 2020, the level of urbanization all over the world has generally increased compared to it in 1980. 
````
